```python
s = "g fmnc wms bgblr rpylqjyrc gr zw fylb. rfyrq ufyr amknsrcpq ypc dmp. bmgle gr gl zw fylb gq glcddgagclr ylb rfyr'q ufw rfgq rcvr gq qm jmle. sqgle qrpgle.kyicrpylq() gq pcamkkclbcb. lmu ynnjw ml rfc spj." 
url = "map"
```

```python
def transformtext(s):
    s2 = [chr(ord(c) + 2) if c != ' ' and c.isalnum() else c for c in s]
    s3 = [chr( ord(c) - 26 ) if ord(c) > 122 else c for c in s2]
    return ''.join(s3)
```

```python
print(transformtext(s) + '\n')
print(transformtext(url))
```

```python
import urllib.request, re
url = r'http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing='
first = str(12345)
```

```python
RE = r'[0-9]+'
prog = re.compile(RE)
```

```python
numbers = [first]
nr = first
while(True)
    html = urllib.request.urlopen(url + nr).read().decode()
    nr = prog.findall(html)[0]
    numbers += [nr]
# ends at 16044
# http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=16044
# says 'Yes. Divide by two and keep going.'
```

```python
nr = str(int(16044/2))
while(True):
    html = urllib.request.urlopen(url + nr).read().decode()
    nr = prog.findall(html)[-1]
    print(nr)
    numbers += [nr]
```
